## Data format function:
Here is the data format function which we use to determine missing items or incorrect entries and replace them.
Also the time in minutes are considered at max 300 for any app.
Any data item exceeding 300 will resukt in the data to be 300 only,if any data item exceeds 600 which is almost 10 hours , it will be omitted and considered as a fault data entry.

```python
def dataformat(obj):
    obj=obj.fillna(0)
    data_object=[]
    for each in obj:
        each=str(each)
        if(each.isnumeric()):
            each=int(each)
        else:
            k=''
            for letter in each:
                if(letter.isnumeric()):
                    k=k+letter
            if(not k):
                each=0
            else:
                each=int(k)
        if(each>300 and each<=600):
            each=300
        if(each>600):
            each=0
        data_object.append(each)
    obj=data_object
    return obj
```

```python
obj=dataformat(df['facebook'])
df['facebook']=obj
df['facebook'].describe()
```
